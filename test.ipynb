```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一些随机数据
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# 创建一个图表，并绘制曲线
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin Function')

# 显示图表
plt.show()
```
